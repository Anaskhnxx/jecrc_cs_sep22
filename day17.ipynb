```python
import matplotlib.pyplot as plt 
import numpy as np 
```

# types of graphs 
1. plot -            line graph , between two paramater 
2. bar -             bar graph of two parameter (optional catagorical_data)
3. hist -            histogram of a bunch of data i.e. frequency of single parameter or two parameter  
4. scattrer -        dot graph between two parameter 
5. boxplot / box -   statistical overview in form of box for a single parameter , mean , outliers , quantiles(25% ,75%) , IQR 
                     (optional - catagorical data) 
6. stem -            more like bar graph , lines instead of bars
7. violinplot -      violin shaped statistical overview graph spread of data (more like hist) single parameter (optional cat-data)
                      

```python
x= np.linspace(0,5,11)
x1 = np.arange(1,11,3)
```

```python
y = x**3
```

```python
plt.plot(x,y)
plt.title('bar graph')
plt.show()
```

```python
plt.bar(x,y)
```

```python
plt.hist(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.title('scatter')
plt.grid()
plt.xlabel('x--->')
plt.ylabel('y--->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.stem(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

```python
plt.violinplot(x)
plt.show
```

y2 = x**4

```python
y2 = x**4
```

```python
plt.plot(x,y)
plt.show()
plt.plot(x,y2)
plt.show()
```

```python
fig  = plt.figure(figsize = (10,5))
plt.plot(x,y, lw=3 , marker ='o' ,markersize = 15, markerfacecolor = 'yellow')
plt.plot(x,y2 , marker = 'v' , markersize = 10 , markerfacecolor = 'black')
plt.legend(['y=x**3' , 'y2=x**4'])
plt.grid()
plt.show()
```
