# graph function :
1. plot                             - simple line plot 
2. plot.area                        - filled version of line plot 
3. plot.bar / plot.barh             - bar graph 
4. plot.box                         - box plot 
5. plot.density                     - density lne (smooth curved)
6. plot.hexbin                      - hexagon version of scatter with density
7. plot.hist                        - histogram
8. plot.kde                         - kernel density estimator (density graph)                       
9. plot.pie                        - pie chart of categorical 
10. plot.scatter                    - scatter graph

```python
fig = plt.figure(figsize = (20,20))
df2.plot(); plt.title('Line Plot')
df2.plot.area(); plt.title('Area Plot') 
df2.plot.bar(); plt.title('Bar Plot') 
df2.plot.box(); plt.title('Box Plot') 
df2.plot.barh(); plt.title('Bar Horizontal Plot') 
df2.plot.density(); plt.title('Density Plot') 
df2.plot.kde(); plt.title('KDE Plot') 
df2.plot.hist(); plt.title('Histogram Plot') 
df2.plot.scatter(x='a', y= 'b' ); plt.title('Scatter Plot') 
df2.plot.hexbin(x='a', y='b' , C='c' , gridsize=6); plt.title('Hexbin Plot') 
```

##### Categories of function
1. categorical plot (hue can be used | comparison with catogorical data)
   1. countplot     (single parameter)
   2. barplot       (one parammeter catogorical and one parameter numerical | calculates mean by default )
   3. boxplot
   4. violinplot
   5. swarmplot
2. distribution plot
   1. distplot
   2. displot
   3. jointplot
   4. pairplot
   5. kdeplot
   6. rugplot
   7. scatterplot
3. matrix plot
   1. heatmap
   2. clustermap
   3. pairplot(regression also draw in pairplot)(same as above me) 
4. regression plot 
   1. implot
   2. regplot
5. grid plots
   1. PairGrid
   2. JointGrid
   3. FacetGrid
   

```python
tips = sns.load_dataset('tips')
sns.set_style('dark')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe(include = 'all')
```

```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex' , 'time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day' , data=tips)
```

```python
sns.countplot(x='day' , data=tips , hue = 'sex')
# hue ------ same as a groupby function 
```

```python
sns.barplot(x = 'day' , y = 'total_bill', data = tips)
```

```python
sns.barplot(x = 'day' , y = 'total_bill', data = tips , estimator=sum)
```

```python
sns.barplot(x = 'day' , y = 'total_bill', data = tips , hue = 'time' , estimator=np.mean)
```

```python
sns.boxplot(x = 'day' , y = 'total_bill', data = tips , hue = 'smoker' )
```

```python
sns.swarmplot(x = 'day' , y = 'total_bill', data = tips , hue = 'smoker' )
```

```python
sns.violinplot(x = 'day' , y = 'total_bill', data = tips , hue = 'smoker' )
```
